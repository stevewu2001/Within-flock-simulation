```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df_vec = [[pd.read_csv(f'Results_vaccinated_chicken_{i}_asym_infect_{j}.csv') for i in [0, 1500, 2250, 2700]] for j in [4.9, 7.0, 14.0, 20.0]]
df_list = df_vec[0]
```

```python
fig, axs = plt.subplots(2, 2, figsize=(12, 8))
vaccinated_chickens = [0, 1500, 2250, 2700]
asymptomatic_infectious_period = [4.9, 7.0, 14.0, 20.0]

for ax, vacc_chicken in zip(axs.flatten(), vaccinated_chickens):
    for df_list, asym_infect in zip(df_vec, asymptomatic_infectious_period):
        df = df_list[vaccinated_chickens.index(vacc_chicken)]
        df_filtered = df[df['Outbreak'] == True]
        ax.hist(df_filtered['Peak Size Chicken'], bins=10, density=True, alpha=0.8, label=f'asymptomatic infectious period = {asym_infect}')
    ax.set_xlabel('Peak Size Chicken')
    ax.set_ylabel('Density')
    ax.set_title(f'vaccinated chicken = {vacc_chicken}')
    ax.legend()

fig.suptitle('Comparison of Peak Number of Infected Chickens for Different vaccination levels and asymptomatic infectious period', fontsize=16)
plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()
```
